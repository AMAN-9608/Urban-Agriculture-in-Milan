```python
# Split the 'info' column into separate substrings
df_combined['Photoperiod'] = df_combined['Photoperiod'].str.split(',').apply(lambda x:[s.strip() for s in x] )

df_combined.head()

# Create binary features using one-hot encoding
# First, we create a Series where each element is a list of substrings from the 'info' column
info_series = df_combined['Photoperiod'].explode()

print(info_series)
print(info_series.index)


# Get dummies (one-hot encoding)
binary_features = pd.get_dummies(info_series).groupby(level=0).max()

df_combined.head()

# # df_combined.reset_index(drop=True, inplace=True)
# # binary_features.reset_index(drop=True, inplace=True)


# Concatenate the binary features with the original DataFrame
df_combined = pd.concat([df_combined.drop(columns=['Photoperiod']), binary_features], axis=1)


print(df_combined)
```

## Model Training

```python
# param grid for random forest
param_grid = {
    'n_estimators': [10,20,30],
    'max_depth': [3,5,10],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4]
}
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split,GridSearchCV
from sklearn.metrics import r2_score,mean_absolute_error,mean_squared_error
from sklearn.preprocessing import StandardScaler

X = df_combined.drop('Yield',axis=1)
y= df_combined['Yield']

X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=10)

sc = StandardScaler()
X_train_scaled = sc.fit_transform(X_train)
X_test_scaled = sc.transform(X_test)

rf = RandomForestRegressor(n_jobs=-1, random_state=10)

grid_search = GridSearchCV(estimator=rf, param_grid=param_grid, cv=5, scoring='neg_mean_squared_error', verbose=2, n_jobs=-1)
grid_search.fit(X_train_scaled, y_train)

# Print the best hyperparameters found
print("Best hyperparameters:", grid_search.best_params_)

# Evaluate the best model on the test set
best_rf = grid_search.best_estimator_
y_pred = best_rf.predict(X_test_scaled)
mse = mean_squared_error(y_test, y_pred)
print("Mean Squared Error on Test Set:", mse)

# rf = RandomForestRegressor(n_estimators=100,random_state=10)
# rf.fit(X_train,y_train)
# y_pred = rf.predict(X_test)

feature_importances = best_rf.feature_importances_

# Create a DataFrame to store feature names and importances
feature_importances_df = pd.DataFrame({
    'Feature': X.columns,
    'Importance': feature_importances
})

# Sort DataFrame by importance score in descending order
feature_importances_df = feature_importances_df.sort_values(by='Importance', ascending=False)


print("Mean Absolute error",mean_absolute_error(y_test,y_pred))
print("r2 score",r2_score(y_test,y_pred))
```

```python
# Find top 20 features
top_20_features = feature_importances_df[:20]
```

## Gradient Boosting Regressor

```python
from sklearn.ensemble import GradientBoostingRegressor
param_grid = {
    'n_estimators': [100, 200, 300],
    'learning_rate': [0.01, 0.05, 0.1],
    'max_depth': [3, 4, 5],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4],
    'subsample': [0.8, 0.9, 1.0]
}

gbr = GradientBoostingRegressor()

grid_search = GridSearchCV(estimator=gbr, param_grid=param_grid, cv=3, n_jobs=-1, verbose=2, scoring='neg_mean_squared_error')

# Fit GridSearchCV to the training data
grid_search.fit(X_train_scaled, y_train)

# Print the best parameters and the corresponding score
print("Best parameters found: ", grid_search.best_params_)
print("Best negative mean squared error: ", grid_search.best_score_)

# Use the best model to make predictions on the test set
best_model = grid_search.best_estimator_
y_pred = best_model.predict(X_test_scaled)

# Evaluate the performance on the test set
mse = mean_squared_error(y_test, y_pred)
print("Mean Squared Error on test set: ", mse)

print("r2_score",r2_score(y_test,y_pred))
```
